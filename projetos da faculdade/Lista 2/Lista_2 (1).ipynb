
4. Faça um programa que peça o valor total de um orçamento (valor igual ou maior do que R$100.00). Então, peça vários valores de serviço até chegar no valor total e mostre na tela o valor de cada serviço que será executado. Note que nem sempre será possível utilizar todo orçamento. Quando o valor do serviço estourar ou consumir todo orçamento, o programa deverá avisar que nenhum outro serviço poderá ser pedido.

```python
orcamento = float(input("Informe o valor do orçamento (mínimo R$100.00): "))

if orcamento < 100:
    print("Orçamento insuficiente.")
else:
    total = 0
    servicos = []

    while True:
        valor = float(input("Informe o valor do serviço: "))

        if total + valor > orcamento:
            print("Orçamento excedido. Nenhum outro serviço poderá ser pedido.")
            break
        else:
            servicos.append(valor)
            total += valor

    print("\nServiços executados:")
    for s in servicos:
        print(f"R${s:.2f}")
```

6. Faça um programa que peça dois números inteiros positivos, base e expoente, calcule e mostre o primeiro número elevado ao segundo número. Não utilize a função ou o operador de potência da linguagem.

```python
base = int(input("Insira um número (base): "))
expoente = int(input("Insira um número (expoente): "))

resultado = 1
contador = 0

while contador < expoente:
    resultado *= base
    contador += 1

print(f"{base} elevado a {expoente} é igual a {resultado}")
```

8. Faça um programa que peça um número inteiro positivo n e mostre os n primeiros números da sequência de Fibonacci.


```python
#8
n1 = 1
n2 = 1
n3 = n2 + n1
n = int(input("Digite um inteiro: "))

if n <= 0:
    print("Quantidade inválida (deveria ser > 0).")
else:
    for i in range(n):
        if i == 0 or i == 1:
            print(1, end=', ')
        elif i > 1 and i < n - 1:
            n3 = n1 + n2
            print(n3, end=", ")
            n1, n2 = n2, n3
        else:
            print(n1 + n2)
```

11. Faça um programa que peça dois números inteiros positivos e mostra o máximo divisor comum (MDC) e o mínimo múltiplo comum (MMC) desses números.

```python
#11
m, n = 6, 4
divisor = 2
mmc = 1
continuar = True
teste_m, teste_n = False, False

print(m, n)

while m != 1 or n != 1:
    while continuar == True:
        if m % divisor == 0:
            m = m // divisor
            teste_m = True

        if n % divisor == 0:
            n = n // divisor
            teste_n = True

        continuar = teste_m or teste_n
        teste_m = False
        teste_n = False

        if continuar == True:
            mmc = mmc * divisor
            print(m, n)
    divisor = divisor + 1
    continuar = True

print(f'mmc = {mmc}')
```

12. Faça um programa que peça a razão e primeiro termo de uma PA (progressão aritmética). Peça um número inteiro positivo n e mostra os n primeiros termos dessa PA.

```python
razao = int(input("Digite a razão da PA: "))
primeiroTermo = int(input("Digite o primeiro termo PA: "))
num = int(input("Digite um número inteiro positivo : "))

if num <= 0:
    print("O número  deve ser inteiro positivo.")
else:
    print("Os termos da progressão aritmética são:")
    for i in range(num):
        termo = primeiroTermo + i * razao
        print(termo)
```

17. Faça um programa que peça o valor presente (capital inicial), a taxa de juros mensais (i) e o período em meses(n). Então, o programa deve calcular e mostrar a evolução do valor futuro (montante) a cada um dos n meses.

```python
capitalInicial = float(input('Digite o valor presente (capital inicial):'))
taxaJM = float(input('Digite a taxa de taxa de juros mensais:'))
periodo = int(input('Digite o período em meses:'))
if periodo <= 0:
    print('O período deve ser maior que zero.')
else:
    print('Evolução do montante mês a mês:')
    montante = capitalInicial
    for mes in range(1, periodo + 1):
        montante *= (1 + taxaJM)
        print(f'Mês {mes}: R$ {montante:.2f}')
```

18. Faça um programa que pede o valor que será depositado todos o meses, o valor da taxa de juros mensal e o total de meses (n) . O programa deverá mostrar na tela o valor do montante e o valor dos juros em cada um dos n meses.

```python
depositoMensal = float(input('Digite o valor que será depositado todos os meses:'))
taxaJuros = float(input('Digite a taxa de juros mensais:'))
totalMeses = int(input('Digite o total de meses:'))

if totalMeses <= 0:
    print('O total de meses deve ser maior que zero.')
else:
    print('Evolução do montante e juros mês a mês:')
    montante = 0
    for mes in range(1, totalMeses + 1):
        juros = montante * taxaJuros
        montante += juros + depositoMensal
        print(f'Mês {mes}: Montante = R$ {montante:.2f}, Juros do mês = R$ {juros:.2f}')
```

19. Faça um programa que pede o número n de provas e as n notas e calcula a média e o desvio padrão dessas notas.

```python
provas = int(input("insira o número de provas:"))
somadasnotas = 0
somaQ = 0
for notas in range(1, provas + 1):
    nota = float(input(f'Digite a nota da prova {notas}: '))
    somadasnotas += nota
    somaQ += (nota ** 2)

media = somadasnotas / provas

desvioPadrao = ((somaQ / provas) - (media ** 2)) ** 0.5

print(f'Média: {media:.2f}')
print(f'Desvio padrão: {desvioPadrao:.2f}')
```
